```python
attack1 = 'スマッシュ'
attack2 = 'ストローク'

for count in range(10):
    if count % 2 == 0: ## countの値を2で割った余りが0であれば次のステートメントを実行
        print(attack1)
    else:              ## それ以外
        print(attack2)
```

```python
## スマッシュとボレーを織り交ぜた攻撃プログラムをリスト内包表記に置き換える
list = [print(attack1) if i % 2 == 0 else print(attack2) for i in range(10)]
```

```python
attack1 = 'ボレー'
attack2 = 'スマッシュ'
attack3 = 'ストローク'
 
for count in range(10):
    if (6 <= count) and (count % 2 == 0): ## countが6以上､かつ2で割った余りが0であるか?
        print(attack1 + '_' + str(count))
    elif (count % 2 == 0):               ## countを2で割った余りが0であるか?
        print(attack2 + '_' + str(count))
    else:                               ## どれにも当てはまらない場合に実行
        print(attack3 + '_' + str(count))
```

```python
import random ## randomモジュールをインポート

attack1 = 'ボレー'
attack2 = 'スマッシュ'
attack3 = 'ロブ'
attack4 = 'ストローク'

for count in range(5):
    x = random.randint(1,10) ## 内部で生成済みのrandomオブジェクトを参照し､1〜10の整数をランダムに生成
    if x <= 3:
        print(attack1 + '_' + str(x))
    elif x >= 4 and x <= 5:
        print(attack2 + '_' + str(x))
    elif x >= 6 and x <= 7:
        print(attack3 + '_' + str(x))
    else:
        print(attack4 + '_' + str(x))
```
