```python
#名前を取得
name = input('お名前は?')
print('%s さん､こんにちは!' % name)

# 入力待ちのメッセージ
prompt = name + 'さん発言をどうぞ'

##　さようならと入力､又は未入力以外は応答を続ける
while 1:
    answer = input(name + 'さん発言をどうぞ')
    if answer == 'さようなら':
        print('バイバイ!')
        break ## ループを抜ける
    ## 未入力ならループ終了
    elif not answer:
        print('.....')
        break ## ループを抜ける
    ## 応答
    print('『{}』なんですね'.format(answer))
```
