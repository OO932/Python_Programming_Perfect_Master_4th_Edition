```python
class MyCoach:
    """鬼コーチクラス
    """
    def __init__(self, max):
        """初期化メソッド

        Args:
            max (init):繰り返しの回数
        """
        sewlf.max = max # インスタンス変数self.maxをmaxパラメータで初期化
        self.count = 0  # インスタンス変数self.countを0で初期化

    def teach(self):
        """指導メソッド
        """
        if self.count < self.max:
            # self.countがself.max未満の時の処理
            print('もっと強く!')
        else:
            # self.countがself.maxに達したら以下を実行
            print('よ-しオッケーだ!')
        
        # カウンターを1増やす
        self.count += 1
```
